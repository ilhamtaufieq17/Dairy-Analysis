## Information about columns

This dataset contains over 92,887 rows and 13 columns. Here is a description of each column:

- Location: Geographical location of the dairy farm.

- Number of Cows: Number of cows in the dairy farm.
- Farm Size: Size of the dairy farm (in square kilometers).
- Date: Date of data recording.
- Product ID: Unique identifier for each dairy product.
- Product Name: Name of the dairy product.
- Brand: Brand associated with the dairy product.
- Quantity (liters/kg): Quantity of dairy product available.
- Price per Unit: Price per unit of dairy product.
- Total Value: Total value of quantity of dairy product available.
- Shelf Life (days): Shelf life of the dairy product in days.
- Storage Conditions: Recommended storage conditions for the dairy product.
- Production Date: Production date for the dairy product.
- Expiry Date: Expiry date for the dairy product.
- Quantity Sold (liters/kg): Quantity of dairy product sold.
- Price per Unit (sold): Price per unit at which the dairy product was sold.
- Estimated Total Revenue (INR): Estimated total revenue generated from the sale of dairy products.
- Customer Location: Location of the customer who purchased the dairy product.
- Sales Channel: Channel through which the dairy product is sold (Retail, Wholesale, Online).
- Quantity in Stock (litres/kg): Quantity of dairy product remaining in stock.
- Minimum Stock Limit (litres/kg): Minimum stock limit for dairy products.
- Reorder Quantity (litres/kg): Recommended quantity to reorder the dairy product.

# <b>Section 4 | Data Exploration</b>

### 1. Which dairy product categories generate the highest revenue?

```python
# Grouping the data by 'Product Name' to find the total revenue generated by each product category
revenue_by_product = df.groupby('Product Name')['Approx. Total Revenue(INR)'].sum()

# Sorting the data to show the highest revenue generating categories at the top
sorted_revenue = revenue_by_product.sort_values(ascending=False)

# Generating colors from the RdYlGn palette
colors = sns.color_palette('RdYlGn', n_colors=len(sorted_revenue))

# Creating a bar chart for visualization
plt.figure(figsize=(10, 6))
ax = sorted_revenue.plot(kind='bar', color=colors)

# Adding annotations on each bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:,.0f}',  # Format number with comma as thousand separator
                (p.get_x() + p.get_width() / 2, height),  # Position of annotation
                ha='center', va='bottom',  # Text alignment
                fontsize=10, color='black')

# Adding title and labels
plt.title('Revenue Generated by Dairy Product Categories')
plt.xlabel('Product Name')
plt.ylabel('Total Revenue (INR)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()
```

From this graph, it can be seen that Curd generates the highest revenue among all the product categories, followed by Butter and Lassi. On the contrary, Ice Cream generates the lowest revenue compared to the other products listed. In conclusion, Curd is the best performing product in terms of revenue, whereas Ice Cream generates relatively lower revenue. This graph provides valuable insights into which products are the most profitable and can be used to focus marketing or distribution efforts more effectively.

### 2. How do customer demographics impact sales?

```python
# Mengelompokkan data berdasarkan 'Customer Location' untuk menjumlahkan 'Approx. Total Revenue(INR)'
revenue_by_location = df.groupby('Customer Location')['Approx. Total Revenue(INR)'].sum().sort_values(ascending=False)

# Mengatur palet warna RdYlGn
colors = sns.color_palette('RdYlGn', n_colors=len(revenue_by_location))

# Membuat plot
plt.figure(figsize=(12, 8))
ax = revenue_by_location.plot(kind='bar', color=colors)

# Menambahkan anotasi pada setiap bar
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{height:,.0f}',  # Format angka dengan koma sebagai pemisah ribuan
                (p.get_x() + p.get_width() / 2, height),  # Posisi anotasi
                ha='center', va='bottom',  # Penempatan anotasi
                fontsize=10, color='black')

# Menambahkan judul dan label
plt.title('Revenue by Customer Location')
plt.xlabel('Customer Location')
plt.ylabel('Total Revenue (INR)')
plt.xticks(rotation=45, ha='right')

# Mengatur layout agar tidak ada elemen yang terpotong
plt.tight_layout()

# Menampilkan plot
plt.show()
```

The graph shown is a bar chart depicting the total revenue (in INR) from various customer locations. Titled “Revenue by Customer Location,” the graph highlights the revenue generated from various regions such as Chandigarh, Delhi, Bihar, Kerala, and others. The X-axis displays the names of the customer locations, while the Y-axis shows the total revenue generated from each location. The bars on the graph vary in color from red to green, reflecting the differences in revenue levels across locations.

Key findings from the graph show that Chandigarh and Delhi are the two highest-revenue customer locations, generating over INR 6 million each. Bihar is also a significant location with revenues above INR 4 million. Other locations such as Kerala, West Bengal, Karnataka, Maharashtra, and Uttar Pradesh have slightly lower but still significant revenues. Haryana emerges as the lowest-revenue location among those shown.

From this graph, it can be concluded that Chandigarh and Delhi are the key markets with the highest revenues, indicating that these regions may require special attention in marketing and distribution strategies. Focusing efforts on these areas can help maintain and expand market share in the most profitable areas.

### 3. How does inventory management affect sales outcomes?

```python

# Filter relevant columns for analysis
data_filtered = data[['Product Name', 'Quantity Sold (liters/kg)', 'Approx. Total Revenue(INR)', 'Quantity in Stock (liters/kg)']]

# Group by Product Name to aggregate data for visualization
grouped_data = data_filtered.groupby('Product Name').sum().reset_index()

# Define the color palette from RdYlGn
cmap = plt.get_cmap('RdYlGn')
num_products = len(grouped_data)
colors = cmap(np.linspace(0.2, 0.8, num_products))  # Generate colors for the number of products

# Create a bar chart to show the relationship between Quantity Sold, Stock, and Revenue by Product
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plotting Quantity Sold with a gradient of colors from RdYlGn
bars = ax1.bar(grouped_data['Product Name'], grouped_data['Quantity Sold (liters/kg)'], color=colors, label='Quantity Sold (liters/kg)')

# Create a secondary y-axis to plot Revenue
ax2 = ax1.twinx()
line, = ax2.plot(grouped_data['Product Name'], grouped_data['Approx. Total Revenue(INR)'], color='black', marker='o', label='Revenue (INR)')

# Labels and title
ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Sold (liters/kg)', color='black')
ax2.set_ylabel('Approx. Total Revenue (INR)', color='black')
ax1.set_title('Relationship between Quantity Sold, Product Name, and Revenue by Product')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Annotate bars with Quantity Sold
for bar in bars:
    height = bar.get_height()
    ax1.annotate(f'{height:,.0f}',  # Format with comma as thousand separator
                 (bar.get_x() + bar.get_width() / 2, height),  # Position above the bar
                 ha='center', va='bottom',  # Center the annotation
                 fontsize=10, color='black')

# Annotate line plot with Revenue below the markers with additional padding
for x, y in zip(grouped_data['Product Name'], grouped_data['Approx. Total Revenue(INR)']):
    ax2.annotate(f'{y:,.0f}',  # Format with comma as thousand separator
                 (x, y),  # Position of the annotation
                 textcoords="offset points",  # Position with offset
                 xytext=(0, -20),  # Padding of 15 points below the marker
                 ha='center', fontsize=10, color='black')

# Show legends
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')


# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

From this graph, it can be seen that “Curd” has the highest volume sold at around 120,415 liters/kg, and the highest revenue at INR 6,746,881. Meanwhile, “Ghee” shows the lowest volume sold at around 98,513 liters/kg, but does not have the lowest revenue. “Ice Cream,” despite selling a fairly high volume at around 104,780 liters/kg, generates the lowest revenue at INR 5,304,634. The graph also shows that there is a significant variation between the sales volume and revenue, indicating a difference in the price per unit between the products. Overall, this graph provides valuable insights into how sales volume and revenue relate to each other for different products, which can help in marketing strategies and business decisions.

### 4. Which sales channels are most successful?

```python
# Ensure 'Approx. Total Revenue(INR)' is numeric, converting if necessary
df['Approx. Total Revenue(INR)'] = pd.to_numeric(df['Approx. Total Revenue(INR)'], errors='coerce')

# Group by Sales Channel to aggregate revenue data
grouped_sales_channel = df.groupby('Sales Channel')['Approx. Total Revenue(INR)'].sum().reset_index()

# Define the color palette from RdYlGn
cmap = plt.get_cmap('RdYlGn')
num_channels = len(grouped_sales_channel)
colors = cmap(np.linspace(0.2, 0.8, num_channels))  # Generate a range of colors

# Create a bar chart for Sales Channel
fig, ax1 = plt.subplots(figsize=(14, 8))

# Bar chart for Sales Channel with colors from RdYlGn
bars = ax1.bar(grouped_sales_channel['Sales Channel'], grouped_sales_channel['Approx. Total Revenue(INR)'], color=colors)
ax1.set_xlabel('Channel')
ax1.set_ylabel('Total Revenue (INR)')
ax1.set_title('Total Revenue by Sales Channel')
ax1.tick_params(axis='x', rotation=45)  # Rotate x-axis labels for better readability

# Annotate bars with total revenue
for bar in bars:
    height = bar.get_height()
    ax1.annotate(f'{height:,.0f}',  # Format with comma as thousand separator
                 (bar.get_x() + bar.get_width() / 2, height),  # Position above the bar
                 ha='center', va='bottom',  # Center the annotation
                 fontsize=10, color='black')

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

This figure shows a bar chart depicting the total revenue generated from various sales channels, namely Online, Retail, and Wholesale. The vertical axis shows the total revenue in INR (Indian Rupees), with a scale ranging from millions of INR. From the graph, it can be seen that the **Retail** channel generated the highest revenue among the three channels, reaching INR 20,863,658. The **Wholesale** channel also generated revenue almost equal to Retail, which was around INR 20,214,596. Meanwhile, the **Online** channel recorded the lowest revenue among the three, at INR 17,656,393. Although Online revenue is lower compared to Retail and Wholesale, the figure is still significant. This graph indicates that Retail and Wholesale are the main sales channels that contribute the most revenue, while the Online channel, although in third place, still plays a significant role in the total revenue.
